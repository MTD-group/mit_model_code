# Test the featurizer on the entire collection of structures

```python
# get the initial dataframe
df = generate_formula_from_structure()
# manually correct the Pymatgen naming from TiPbO3 to PbTiO3
df.Compound[df.Compound == "TiPbO3"] = "PbTiO3"
```

```python
df
```

Some structures need to be made into supercells of 2a, 2b, 2c in order for the element distances to be featurized correctly

```python
modify_index_lst = [0, 4, 5, 12, 14, 35, 36, 38, 47, 52, 53, 71, 73, 74, 82, 83, 84, 102, 104, 105, 106, 107, 112, 116, 117, 118, 126, 137, 142, 
                    147, 149, 160, 164, 172, 177, 180, 187, 188, 190, 195, 196, 197, 198, 199, 206, 215, 226, 231, 240, 246, 248, 250, 251, 
                    252, 254, 255, 256, 258, 259, 260, 261, 265, 270, 272, 274, 286, 292, 294, 340, 342, 344, 364, 366, 367, 372, 375, 401]
df.iloc[modify_index_lst].Compound.to_list()
```

```python
for i in modify_index_lst:
    # retrieve the structure
    struct_to_be_modified = df.at[i, "structure"]
    # make a supercell of 2a, 2b, 2c (in-place modification)
    struct_to_be_modified.make_supercell([2, 2, 2])
```

## Generate compositional, structural and handbuilt features

```python
df = composition_featurizer(df)
df = structure_featurizer(df)
df = handbuilt_featurizer(df)
```
